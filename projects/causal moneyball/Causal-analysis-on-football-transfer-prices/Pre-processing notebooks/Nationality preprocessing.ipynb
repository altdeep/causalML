```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from tqdm import tqdm
import pycountry_convert as pc
import pycountry

warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_csv('Causal-analysis-on-football-transfer-prices-data-preprocessing/data/Data for different model combinations/transfers111.csv')
df.head()

df['Groups'] = 0

for i in range(len(df)):
    country_code = pc.country_name_to_country_alpha2(df.N[i], cn_name_format="default")
    continent_name = pc.country_alpha2_to_continent_code(country_code)
    df['Groups'][i] = continent_name
df
```
